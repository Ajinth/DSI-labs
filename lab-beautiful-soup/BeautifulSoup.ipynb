````python
foods = []

for r in restaurants:
    print('Scraping {}'.format(r['Name']))
    res_rest = requests.get('http://chubbygrub.com/restaurants/{}'.format(r['slug']))
    r_soup = BeautifulSoup(res_rest.content, 'lxml')
    table = r_soup.find('table',{'id':'items'})
    for row in table.find('tbody').find_all('tr'):
        cells = row.find_all('td')
        food = {}
        food['restaurant'] = r['Name']
        food['name'] = cells[0].text
        food['category'] = cells[1].text.strip()
        food['calories'] = cells[2].text
        food['fat'] = cells[3].text
        food['carbs'] = cells[4].text
        foods.append(food)
````
